## 2. Chat Models


#### Input >> Messages 

#### Output >> Messages

```python
from langchain_aws import ChatBedrockConverse,BedrockEmbeddings
llm=ChatBedrockConverse(model='cohere.command-r-plus-v1:0', #amazon.nova-lite-v1:0
                        aws_access_key_id='',
                        aws_secret_access_key='',
                       region_name='us-east-1',max_tokens=200)
```

```python
from rich import print
```

```python
system_prompt = (
    "You are an assistant that helps write polite business emails."

)
user_input=input()
messages = [
            ("system", system_prompt),
            ("human", user_input),
        ]

response = llm.invoke(messages)
print(response)
```

```python
from rich import print
from rich.prompt import Prompt

# Define system prompt once
system_prompt = (
    "You are an assistant that helps write polite business emails."

)

while True:
    current_prompt = Prompt.ask("[bold red]User[/bold red]")
    if current_prompt.lower() == 'exit':
        print(f"[bold blue]Thanks for the conversation.[/bold blue]")
        break
    else:
        # Construct messages as (role, content) tuples
        messages = [
            ("system", system_prompt),
            ("human", current_prompt),
        ]

        response = llm.invoke(messages)
        print(f"[bold green]AI[/bold green]: {response.content}")
```
