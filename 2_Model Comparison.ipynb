### Create multiple LLM instances with different models

```python
from langchain_aws import ChatBedrockConverse,BedrockEmbeddings
llm_2 = ChatBedrockConverse(model='amazon.nova-lite-v1:0',
                        aws_access_key_id='',
                        aws_secret_access_key='',
                       region_name='us-east-1',max_tokens=200)
```

```python
from langchain_aws import ChatBedrockConverse,BedrockEmbeddings
llm_3 = ChatBedrockConverse(model="anthropic.claude-3-sonnet-20240229-v1:0",  #'cohere.command-r-plus-v1:0', #amazon.nova-lite-v1:0
                        aws_access_key_id='',
                        aws_secret_access_key='',
                       region_name='us-east-1',max_tokens=200)
```

```python
llms=[llm_1,llm_2,llm_3]
```

```python
from langchain_classic.model_laboratory import ModelLaboratory
```

```python
model_lab = ModelLaboratory.from_llms(llms)
model_lab.compare("What is the name of the LLM architecture described in the paper 'Attention is all you need'")
```

All three models correctly identified the Transformer architecture, but they varied significantly in their depth and formatting:

Anthropic Claude 3 Sonnet provided the most structured and technical breakdown, using bullet points to explain specific components like multi-head attention and positional encodings. In contrast, Cohere Command R+ and Amazon Nova-Lite delivered more concise, prose-based summaries that focused on the historical shift from traditional recurrent layers to self-attention.
