Prompt templates help structure user input and parameters into clear instructions for language models. This ensures the model responds in a contextually appropriate and coherent way.

##### A prompt template takes a dictionary as input. Each key in the dictionary represents a variable to be filled in the template.

##### It returns a PromptValue, which:
   1.  Can be sent directly to an LLM or ChatModel.
   2.  Can be converted to either a string (for standard LLMs) or a list of messages (for chat models).
   3.  It simplifies switching between string-based and message-based model inputs.


<font color=Blue><b>There are different prompt templates: <br>


### StringPromptTemplate

    These prompt templates format a single string and are typically used for simple inputs.

```python
from langchain_core.prompts import PromptTemplate

prompt_template = PromptTemplate.from_template("Write a short note on the following topic: {topic}")

prompt_template.invoke({"topic": "Evolution"})
```

### ChatPromptTemplates

    These prompt templates format a list of messages, each defined by its own template within the overall structure.

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate([
    ("system", "You are a helpful finace assistant"),
    ("user", "Descibe the following finance term in 1 line: {finance_term}")
])

prompt_template.invoke({"finance_term": "repo_rate"})
```

```python
from langchain_core.prompts import ChatPromptTemplate

prompt_template = ChatPromptTemplate([
    ("system", "You are a helpful finace assistant"),
    ("user", "Descibe the following finance term in 1 line: {finance_term}")
])

prompt_template.invoke({"finance_term": "repo_rate"})
```

### MessagesPlaceholder 

    MessagesPlaceholder allows inserting a list of messages into a specific position within a prompt.
    
    Unlike predefined message templates, it lets users dynamically pass a message list to be placed where needed.

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage
```

```python
prompt_template = ChatPromptTemplate([
    ("system", "You are a helpful health_care assistant"),
    ("placeholder", "{msgs}")
])
prompt_template.invoke({"msgs": [HumanMessage(content="hi!")]})
```

### Prompt Template with Multiple Placeholders

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.messages import HumanMessage, AIMessage

# Define a prompt with a system message and a placeholder for previous messages
prompt_template = ChatPromptTemplate([
    ("system", "You are a friendly banking assistant."),
    ("placeholder", "{chat_history}"),
    ("placeholder", "{chat_history2}"),
    ("human", "{question}")
])

# Simulated past conversation
chat_history = [
    HumanMessage(content="What's my name?"),
    AIMessage(content="Your name is Raj")
]
chat_history2 = [
    HumanMessage(content="What's my account balance?"),
    AIMessage(content="Your current balance is $3,500.")
]
# New question
question = "Can you help me transfer $1500 to my savings?"

# Invoke the prompt
formatted = prompt_template.invoke({
    "chat_history": chat_history,
    "question": question,
    "chat_history2": chat_history2,
})

# Output the list of messages sent to the LLM
print(formatted)
```
