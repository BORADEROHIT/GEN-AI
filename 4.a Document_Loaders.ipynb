## Document Loader Demos

DocumentLoaders load data into the standard LangChain Document format.



### Document Loaders in LangChain

| **File Type**       | **LangChain Loader Class**                                            |
| ------------------- | --------------------------------------------------------------------- |
| PDF                 | `langchain_community.document_loaders.PyPDFLoader`                    |
| PDF (alt)           | `langchain_community.document_loaders.PDFMinerLoader`                 |
| HTML / Web Pages    | `langchain_community.document_loaders.WebBaseLoader`                  |
| CSV                 | `langchain_community.document_loaders.CSVLoader`                      |
| SQL Databases       | `langchain_community.document_loaders.SQLDatabaseLoader`              |
| Markdown            | `langchain_community.document_loaders.UnstructuredMarkdownLoader`     |
| JSON                | `langchain_community.document_loaders.JSONLoader`                     |
| JSONL (JSON Lines)  | `langchain_community.document_loaders.JSONLinesLoader`                |
| DOCX (Word)         | `langchain_community.document_loaders.UnstructuredWordDocumentLoader` |
| PowerPoint (PPTX)   | `langchain_community.document_loaders.UnstructuredPowerPointLoader`   |
| YouTube Transcripts | `langchain_community.document_loaders.YouTubeLoader`                  |
| Notion Pages        | `langchain_community.document_loaders.NotionDBLoader`                 |
| XML                 | `langchain_community.document_loaders.UnstructuredXMLLoader`          |
| TXT / Text Files    | `langchain_community.document_loaders.TextLoader`                     |
| Email (.eml)        | `langchain_community.document_loaders.UnstructuredEmailLoader`        |
| ExcelLoader        | `from langchain_community.document_loaders import UnstructuredExcelLoader`|




For more details visit the website : https://docs.langchain.com/oss/python/integrations/document_loaders

## The following code examples demonstrate a few of the popular document loaders.

## 1. PDF Example – Using PyPDFLoader

```python
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("IntroToCloud.pdf")
documents = loader.load()
documents
```

```python
documents[0].page_content
```

## 2. CSV Example – Using CSVLoader

```python
from langchain_community.document_loaders import CSVLoader

loader = CSVLoader(file_path="advertising_data.csv")
documents = loader.load()
print(documents)
```

```python
print(documents[0].page_content)
```

## 3. TXT Example – Using TextLoader

```python
from langchain_community.document_loaders import TextLoader

loader = TextLoader("notes.txt")
documents = loader.load()
print(documents[0].page_content)
```

## 5. XML Example – Using UnstructuredXMLLoader

```python
from langchain_community.document_loaders import UnstructuredXMLLoader

loader = UnstructuredXMLLoader("research_project.xml")
documents = loader.load()
print(documents[0].page_content)
```
